```python
%matplotlib inline
import math
def gradient_descent(x,y):
    m_curr = b_curr = 0
    rate = 0.01
    n = len(x)
    prev_cost = 1000000
    stopping_threshold = 1e-6 
    #plt.scatter(x,y,color='red',marker='+',linewidth='5')
    for i in range(10000):
        y_predicted = m_curr * x + b_curr
        #cost = np.sum([val**2 for val in (y-y_predicted)])*(1/n)
        cost = np.sum((y-y_predicted)**2)*(1/n)
        print ("m = {} , b= {}, cost ={} at iteration {}".format(m_curr,b_curr,cost,i))
        
        if(math.isclose(prev_cost,cost,rel_tol=1e-20)):
            break;
        #if (abs(prev_cost-cost)<=stopping_threshold):
        #    break
        prev_cost = cost 
        #previous_cost = current_cost
        
    # plt.plot(x,y_predicted,color='green')
        md = -(2/n)*sum(x*(y-y_predicted))
        yd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - rate * md
        b_curr = b_curr - rate * yd
     
        
```
